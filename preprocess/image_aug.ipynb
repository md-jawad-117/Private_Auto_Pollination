```python
import os
from PIL import Image, ImageEnhance

def load_and_augment_images(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Process each image in the input folder
    for image_name in os.listdir(input_folder):
        if image_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            image_path = os.path.join(input_folder, image_name)
            image = Image.open(image_path)

            # # Rotate 90 degrees
            # rotated_90 = image.rotate(90)
            # rotated_90.save(os.path.join(output_folder, f"90_{image_name}"))

            # # Rotate 180 degrees
            # rotated_180 = image.rotate(180)
            # rotated_180.save(os.path.join(output_folder, f"180_{image_name}"))
            
            # # Rotate 270 degrees
            # rotated_270 = image.rotate(270)
            # rotated_270.save(os.path.join(output_folder, f"270_{image_name}"))

            # Darken the image
            enhancer = ImageEnhance.Brightness(image)
            darkened_image = enhancer.enhance(0.5)  # Decrease brightness to darken the image
            darkened_image.save(os.path.join(output_folder, f"dark_{image_name}"))

# Example usage
input_folder = 'E:/Research_Papers/Autonomous Pollination/Dataset/aug/images'  # Replace with your input folder path
output_folder = 'E:/Research_Papers/Autonomous Pollination/Dataset/aug/images_dark' # Replace with your output folder path
load_and_augment_images(input_folder, output_folder)
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

def augment_and_display(image_path):
    # Load the image
    image = Image.open(image_path)

    # Rotate 90 degrees
    rotated_90 = image.rotate(90)

    # Rotate 180 degrees
    rotated_180 = image.rotate(180)

    # Display the images side by side
    fig, axes = plt.subplots(1, 3, figsize=(12, 6))

    axes[0].imshow(image)
    axes[0].set_title('Original')

    axes[1].imshow(rotated_90)
    axes[1].set_title('Rotated 90°')

    axes[2].imshow(rotated_180)
    axes[2].set_title('Rotated 180°')

    # Hide the axes
    for ax in axes:
        ax.axis('on')
    plt.savefig("E:/Research_Papers/weed_detect/images/aug.jpg")
    plt.show()


# Example usage
input_image_path = 'E:/Research_Papers/weed_detect/Dataset/Final/train/images/pp.jpg'  # Replace with the path to your input image
augment_and_display(input_image_path)
```
